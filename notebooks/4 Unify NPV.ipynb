```python
# Import libraries
import pandas as pd
from investing.config.file_paths import processed_data, output_data
```

## Configure

```python
# File names
input_file_fcf = "npv_fcf.csv"
input_file_div = "npv_dividends.csv"
output_file = "npv.csv"
```

## Load Data

```python
# Open the files
df_fcf = pd.read_csv(processed_data / input_file_fcf)
df_div = pd.read_csv(processed_data / input_file_div)
```

```python
# Display data
df_fcf
```

```python
# Display data
df_div
```

### Concatenate Data

```python
# Concatenate dataframes
df=pd.concat([df_fcf, df_div])
df
```

## Save

```python
# Write back df to csv
df.to_csv(output_data / output_file, index=False)
```
